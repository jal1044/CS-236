## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path 
show - Show all items 
drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. 
Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. 
Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit 
Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import random

cont = "y"

def get_command(x):
    if x == "walk":
        walk()
    elif x == "show":
        show()
    elif x == "drop":
        drop()
    elif x == "exit":
        global cont
        cont = "n"
    else:
        print("Invalid input.  Please try again.")
        get_command(input("Command: "))

def walk():
    with open("wizard_all_items.txt") as fItems:
        items = fItems.readlines()
    x = random.randrange(0,len(items))
    print("While walking down a path, you see " + items[x]  + ".")
    grab = input("Do you want to grab it? (y/n): ")
    if grab.lower() == "y":
        with open("wizard_inventory.txt") as fInventory:
            inventory = fInventory.readlines()
        if len(inventory) >= 4:
            with open("wizard_inventory.txt", "w") as fInventory:
                for i in inventory:
                    fInventory.write(i + "\n")
            print("You picked up " + items[x] + ".\n")
        else:
            print("You can't carry any more items. Drop something first.\n")
    elif  grab.lower() == "n":
        print("You pass by and continue walking.\n")

def show():
    with open("wizard_inventory.txt") as fInventory:
        inventory = fInventory.readlines()
    for i in range(0,len(inventory),1):
        print(str(i+1) + ". " + inventory[i])

def drop():
    with open("wizard_inventory.txt") as fInventory:
        inventory = fInventory.readlines()
    try:
        slot = int(input("Inventory Slot Number: "))
    except:
        print("Invalid input.  Please try again.")
        drop()
    if slot < 0 or slot > (len(inventory)+1):
        print("Invalid input.  Please try again.")
        drop()
    else:
        dropped = inventory.pop(slot-1)
        with open("wizard_inventory.txt", "w") as fInventory:
                for i in inventory:
                    fInventory.write(i + "\n")
        print("You dropped " + dropped + ".")
        

def main():
    print("The Wizard Inventroy Program\n\nCOMMAND MENU\nwalk - Walk down the path\nshow - Show all items\ndrop - Drop an item\nexit - Exit program")
    global cont
    while cont == "y":
        print()
        get_command(input("Command: "))
    print("Bye!")

if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:
```powershell

COMMAND MENU

yearly	- View yearly sumary

```

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import csv

cont = "y"

def get_command(x):
    if x == "monthly":
        monthly()
    elif x == "yearly":
        yearly()
    elif x == "edit":
        edit()
    elif x == "exit":
        global cont 
        cont = "n"
    else:
        print("Invalid input.  Please try again.")
        get_command(input("Command: "))

def monthly():
    with open("monthly_sales.csv", newline="") as sales:
        reader = csv.reader(sales)
        for row in reader:
            print(row[0] + " - " + (str(row[1])))
            
def yearly():
    with open("monthly_sales.csv", "r", newline="") as sales:
        reader = csv.reader(sales)
        total = 0
        months = 0
        for row in reader:
            months = months+1
            total = total+int(row[1])
        print("Yearly total:	" + str(total))
        print("Monthly average:	" + str(round((total/months),2)))
        
def edit():
    month = input("Three-letter Month: ")
    rows = []
    valid = False
    with open("monthly_sales.csv", newline="") as sales:
        reader = csv.reader(sales)
        for row in reader:
            rows.append(row)
    for i in range(len(rows)):
        if rows[i][0] == month:
            valid = True
            sale = input("Sales Amount: ")
            rows[i][1] = sale
    if valid == False:
        print("Invalid input.  Please try again.")
        edit()
    with open("monthly_sales.csv", "w", newline="") as sales:
        writer = csv.writer(sales)
        writer.writerows(rows)
    print("Sales amount for " + month + " was modified.")

def main():
    print("Monthly Sales program\n\nCOMMAND MENU\nmonthly - View monthly sales\nyearly	- View yearly sumary\nedit	- Edit sales for a month\nexit	- Exit program")
    global cont
    while cont == "y":
        print()
        get_command(input("Command: "))
    print("Bye!")

if __name__ == "__main__":
    main()
````
