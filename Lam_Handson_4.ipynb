## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
def getInt():
    num = int(input("Please enter an integer between 1 and 5000: "))
    if num < 2 or num > 4999:
        print("Invalid integer. Please try again.")
        getInt()
    return num

def getFactors(num):
    factors=[]
    for i in range (1,num+1,1):
        x = num%i
        if x == 0:
            factors.append(i)
    return factors
            

def main():
    print("Prime Number Checker\n")
    cont = "y"
    while cont == "y":
        num = getInt()
        factors = getFactors(num)
        if len(factors) == 2:
            print(str(num) + " is a prime number.")
        else:
            print(str(num) + " is NOT a prime number.\nIt has " + str(len(factors)) + " factors:" + str(factors))
        cont = input("\nTry again? (y/n): ")
        cont = cont.lower()
        print("")
    print("Bye!")

if __name__ == "__main__":
    main()
````

````python
contacts = [["Guido van Rossum", "guido@rossum.com", "123-456-7890"], ["Eric Idle", "eric@ericidle.com", "+44 20 7946 0958"]]
cont = "y"

def get_command(x):
    global cont
    if x == "list":
        show_list()
    elif x == "view":
        view()
    elif x == "add":
        add()
    elif x == "del":
        delete()
    elif x == "exit":
        cont = "n"
    else:
        print("Invalid input.  Please try again.")
        get_command(input("Command: "))

def show_list():
    global contacts
    for i in range (0,len(contacts),1):
        print(str(i+1) + ".\t" + str(contacts[i][0]))

def view():
    global contacts
    x = get_contact_num()
    print("Name:\t" + contacts[x-1][0])
    print("Email:\t" + contacts[x-1][1])
    print("Phone:\t" + contacts[x-1][2])

def add():
    global contacts
    contact = []
    contact.append(input("Name: "))
    contact.append(input("Email: "))
    contact.append(input("Phone: "))
    contacts.append(contact)
    print(contact[0] + " was added.")

def delete():
    global contacts
    x = get_contact_num()
    deleted = contacts.pop(x-1)
    print(deleted[0] + " was deleted.")

def get_contact_num():
    x = 0
    while x == 0:
        try:
            x = int(input("Number: "))
        except:
            print("Invalid input.  Please try again.")
            x = 0
    if x < 1 or x > len(contacts):
        print("Invalid input.  Please try again.")
        x = get_contact_num()
    return x

def main():
    global contacts
    global cont
    print("Contact Manager\n")

    print("COMMAND MENU\nlist - Display all contacts\nview - View a contact\nadd - Add a contact\ndel - Delete a contact\nexit - Exit program\n")
    while cont == "y":
        get_command(input("Command: "))
        print("\n")
    print("Bye!")

if __name__ == "__main__":
    main()
````
